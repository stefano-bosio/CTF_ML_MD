# Basic introduction to Python 

In Python, there are several built-in data types that can be used to store and manipulate data. Some of the most common data types are:

- int: This data type is used to store integer values. For example, 10, -3, 0 are all examples of integers.

- float: This data type is used to store floating-point numbers. For example, 3.14, 0.1, -2.718 are all examples of floating-point numbers.

- bool: This data type is used to store boolean values, either True or False.

- str: This data type is used to store text data, also known as strings. Strings are enclosed in quotation marks, either single or double quotes. For example, "hello", 'world'.

SOme data structures: 

- list: Used to store a collection of values (could be both numbers of strins). Lists are enclosed in square brackets and the values are separated by commas. For example, [1, 2, 3], ['ASP', 'ASN', 'TYR'].

- tuple: Similar to lists, but the values in a tuple cannot be changed once they are assigned. Tuples are enclosed in parentheses and the values are separated by commas. For example, (1, 2, 3), ('ASP', 'ASN', 'TYR').

- dict: Used to store values in a key-value format. Dictionaries are enclosed in curly braces and the values are separated by colons. For example, {'Atomicsymbol': 'C', 'mass': 12}.

- arrays: Used to store a collection of items, similar to lists and tuples. However, arrays have some advantages over lists and tuples in terms of performance and memory usage, especially when working with large amounts of numerical data. Can be used to perform operations such as indexing, slicing, and iteration, as well as mathematical operations such as addition and multiplication.

# Matrix products (recap + examples)

Matrix product, also known as matrix multiplication, is a mathematical operation that takes two matrices as inputs and produces a third matrix as output. In order for two matrices to be multiplied, the inner dimensions must match.

For example, if you have a matrix A of size (2, 4) and a matrix B of size (4, 4), you can multiply A and B (A * B), but not B and A (B * A). This requirement is due to the definition of a dot product:

$ C(i,j) = \sum_{k=1}^N A(i,k) \cdot B(k,j)$

where the (i, j)th element of C is the scalar-scalar product between the ith row of A and the jth column of B, with k serving as an index.

On the other hand, element-wise multiplication, also known as Hadamard product, is a type of matrix product where each element of the resulting matrix is the product of the corresponding elements from the input matrices.

$C(i,j) = A(i,j) * B(i,j)$

Note that for element-wise multiplication, both matrices must be of the same size.

The "asterisk" $C(i,j) = A(i,j) * B(i,j)$) means element-by-element (element-wise) multiplication while the "dot" ($C(i,j) = \sum_{k=1}^N A(i,k) \cdot B(k,j)$) means dot product, or inner product, i.e. matrix multiplication.

## Dot product (or "inner product")

The inner product, also known as dot product, is an important type of multiplication that can be performed on vectors, which returns a scalar value. The inner product is defined as:

$\mathbf{a} \cdot \mathbf{b} = \mathbf{a}^T\mathbf{b} = \sum_{i=1}^N a_i b_i $


where a and b are vectors, N is the number of elements in the vectors, a_i and b_i are the elements of the vectors, and T represents the transpose operation.

The result of the inner product can be used to calculate the magnitude of the vectors and the cosine of the angle between them. The magnitude of the inner product is given by:


$| a \cdot b | = |a||b| cos \theta_{ab} $

where |a| and |b| are the magnitudes of the vectors a and b, and θ_ab is the angle between the vectors.

Additionally, the cosine of the angle between the vectors can be calculated using the inner product as:

$cos \theta_{ab} = \frac {a^T b}{|a||b|}$

Note that the inner product notation with a transpose implies that the vectors are assumed to be columns by default, and the transposition operation takes a vector as a column and returns it as a row. This allows for element-wise multiplication to occur.

## Outer product 

```python
a = np.array([1,2])
b = np.array([3,4])
```

Element-wise product:  $C(i,j) = A(i,j) * B(i,j)$

```python
a*b
```

Inner (dot) product: $C(i,j) = \sum_{k=1}^N A(i,k) \cdot B(k,j)$

```python
np.dot(a,b)
```

Outer product: $C(i,j) = A(i) B(j)$ 

so that Inner product is the sum product over $i$ in $A(i)B(i)$

```python
np.outer(a,b)
```

Note that its diagonal is the element-wise product:

```python
np.diag(np.outer(a,b))
```

And the trace gives the inner product:

```python
np.trace(np.outer(a,b))
```

## Sampling datas from different distributions

### Random floats in the half-open interval [0.0, 1.0)

```python
R = np.random.random(1000)  
R
```

```python
R.mean()
```

```python
R.var()
```

```python
nbins=30
count, bins, ignored = plt.hist(R, nbins, density=True)
plt.show()
```

### Gaussian distribution

```python
mu=0
sigma=1
npoints=1000
```

```python
G=np.random.default_rng().normal(mu,sigma,npoints)
```

```python
G.mean()  
```

```python
G.var()  
```

Let's try to plot these point and compare them with the corresponding gaussian distribution

```python
nbins=30
count, bins, ignored = plt.hist(G, nbins, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
         np.exp( - (bins - mu)**2 / (2 * sigma**2) ),linewidth=2, color='r')
plt.show()
```

### Poisson distribution

```python
lam =5 # Expected number of events occurring in a fixed-time interval, must be >= 0
```

```python
npoints=10000
```

```python
P=np.random.default_rng().poisson(lam,npoints)
```

```python
nbins=15
count, bins, ignored = plt.hist(P, nbins , density=True)
plt.show()
```
